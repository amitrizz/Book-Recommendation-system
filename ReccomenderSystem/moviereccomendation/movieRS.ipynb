# Now Preprocess The Data 

```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies =movies.merge(credits,on="title")
```

```python
movies.info()
```

```python
#genres    type of movie like action comedy adventure documentry     
#id        movie id to find information about movie 
#keywords  culture clash spy dc comics based on novel 
#title     name of movie
#overview  what happen in movie
#cast      actors
#crew      director
movies = movies[["movie_id","title","overview","keywords","genres","cast","crew"]]
# print(movies["cast"])
```

```python
movies.head(1)
```

```python
movies.isnull().sum()   # checking wheather any data is null or not
```

```python
movies.dropna(inplace=True)    # since null data is very small so droping these movies data from our set
movies.isnull().sum()   
```

```python
movies.duplicated().sum() # to check Any duplicate data is present or not in data set
# movies.drop_duplicates(inplace=True)  # used to delete duplicate data
```

```python
# since it genrate string object so we need to convert string object to normal object 
obj = movies.iloc[0].genres
print(obj)
```

```python
import json
obj = json.loads(movies.iloc[0].genres)
print(obj)
```

```python
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# ["action","Adventure","Fantasy","science Fiction"]  # need data in this form 
# convert function takes string list obj and return list of obj
def fatch_movieType(obj):
  arr=[]
  for i in json.loads(obj):
    arr.append(i["name"]);
  return arr;  
```

```python
movies["genres"]=movies["genres"].apply(fatch_movieType)
```

```python
movies['keywords']=movies['keywords'].apply(fatch_movieType)
```

```python
movies.head(1)
```

```python
def fatch_topCrew(obj):
  arr=[]
  counter=0;
  for i in json.loads(obj):
    if counter!=3:
      arr.append(i["name"]);
      counter =counter+1
    else:
      break  
  return arr; 
```

```python
movies['cast']=movies['cast'].apply(fatch_topCrew)
```

```python
movies['cast']
```

```python
def fatch_director(obj):
  arr=[]
  for i in json.loads(obj):
    if i['job']=="Director":
      arr.append(i["name"]); 
      break;
  return arr; 
```

```python
movies['crew']=movies['crew'].apply(fatch_director)
```

```python
movies['crew']
```

```python
#since overview is a string so we need to convert that string into list so we can cancate all the list 
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview']
movies.head(1)
```

```python
# surname of two author maybe same or last name maybe same to remove
# example like sam loda  sam bsdk here two here system can not identify which one is reccomend
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
# now concate all last 5 columns togather to make a single column 
movies['tags']=movies['overview']+movies['keywords']+movies['cast']+movies['crew']+movies['genres']
movies.head(1)
```

```python
#now we need only three column movie_id title tags
newmovie_data =movies[['movie_id','title','tags']]
newmovie_data.head(1)
```

```python
newmovie_data['tags']=newmovie_data['tags'].apply(lambda x:" ".join(x))
```

```python
newmovie_data['tags'][0]
```

```python
# convert into small letters to no code mismatching
newmovie_data['tags']=newmovie_data['tags'].apply(lambda x:x.lower())
```

```python
newmovie_data['tags'][0]
```

```python
# since some of the word like action or actions can be occure.these word store in diffrent  
# so we need to replace all word with same like action and actons are replace as action
# here we use that library to perform that task
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def strm(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)  
```

```python
newmovie_data['tags']=newmovie_data['tags'].apply(strm)
```

```python
newmovie_data['tags'][0]
```

# Building A model


```python
#since we are found some similarity between text so get some related data to user 
#so we use vectorization in tags so when we search for particular thing/movie/actor than related 
# data will be showed on as roccomend to new user  this whole thing is doen by a library call sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(newmovie_data['tags']).toarray()
cv.get_feature_names_out()   # use to show which words are use here
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
#Now to find the similarity we use cosin similarity (agnle between two movies)
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity[0]
```

```python
# Now we have to create a function in which user input a movie and 
# we have to give related movies that particular movie
# function approch look like that we got a movie now using that movie 
# find the index of that movie and pass to similarity matrix of that movie
# that we sort that index and retrun top indexs 
# if we directly apply sort than we lost the movie index so before sort pair of index with similarity(index,similarity)
def reccomend(movie):
  movie_index=newmovie_data[newmovie_data['title']==movie].index[0]
  distance =similarity[movie_index]
  topfive_index =sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
  fatchmovies=[]
  for i in topfive_index:
    fatchmovies.append(newmovie_data['title'][i[0]])
  return fatchmovies
  
reccomend("Avatar")     
```
